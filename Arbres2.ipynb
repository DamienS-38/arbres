En utilisant la structure de données Dict compter le nombre de représentants de chaque genre botanique : Quel est le genre botanique le plus représenté ?

```python
dict={}
counter=0
list=[]
genre_counter = {}
for ligne in data[1:]:
    #Condition ville= Grenoble et genre != Vide
    if ligne[30]=='Ville de Grenoble' and ligne[14]!='' and ligne[14]!='inconnu':
        #print("ici")
        #Affiche les genres botaniques prèsent à Grenoble
        #print((ligne[14]))
        
        #enregistre les genres dans un dictionnaire
        #dict['genre']=ligne[13]
        #dict['espece']=ligne[14]
        genre = ligne[13]
     
        # Update the counter for the genre
        if genre in genre_counter:
            genre_counter[genre] += 1
        else:
            genre_counter[genre] = 1
     
     
       #enregistre les genres dans une liste
        entry = {
            'genre': ligne[13],
            'espece': ligne[14]
        }
        list.append(entry)
        dict[counter]=entry
        counter +=1

#Affiche les genres botaniques uniques prèsent à Grenoble
print(dict)
print(list)
print(genre_counter)

genre_counter=sorted(genre_counter)
max_genre=max(genre_counter)
print(f"genre max: "+max_genre)
```
