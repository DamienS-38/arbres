```python
#Quelle est la plus ancienne année de plantation recensée dans ce dataset ? --> min()
cmb=[]
for ligne in data[1:]:
    #affiche toute les lignes du fichier de la colonne (sous_categorie_desc)
    #print (ligne[8])
   
    #Enregistre le nombre de caractère des cellules de la colonne--> sous_categorie_desc
    x=ligne[19]
    if x!='':
        cmb.append(x)

print(cmb)

cmb_complet= set(cmb)
print(cmb_complet)

#Trie par ordre croissant
cmb_range=sorted(cmb_complet)
print(cmb_range)

min_annee=min(cmb_range)
print(min_annee)
```

Quelle est la plus récente année de plantation recensée dans ce dataset ?

```python
cmb=[]
for ligne in data[1:]:
    #affiche toute les lignes du fichier de la colonne (sous_categorie_desc)
    #print (ligne[8])
   
    #Enregistre le nombre de caractère des cellules de la colonne--> sous_categorie_desc
    x=ligne[19]
    if x!='':
        cmb.append(x)

print(cmb)

cmb_complet= set(cmb)
print(cmb_complet)

#Trie par ordre croissant
cmb_range=sorted(cmb_complet)
print(cmb_range)

#Trie par ordre décroissant
cmb_range=sorted(cmb_complet,reverse=True)
print(cmb_range)

max_annee=max(cmb_range)
print(max_annee)
```
